# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time
class kumanda():
    def __init__(self,tvdurum="Kapalı",tvses=0,kanallistesi=["Trt"],kanal="Trt"):
        self.tvdurum=tvdurum
        self.tvses=tvses
        self.kanallistesi=kanallistesi
        self.kanal=kanal
    def tvac(self):
        if self.tvdurum=="Açık":
            print("Televizyonunuz açık")
        else:
            print("TV açılıyor")
            self.tvdurum="Açık"
    def tvkapat(self):
        if self.tvdurum=="Kapalı":
            print("Tv zaten kapalı")
        else:
            print("Tv kapanıyor")
            self.tvdurum=="Kapalı"
    def sesayarları(self):
        while True:
            cevap = input("<,>,çıkış")
            if cevap == "<":
                self.tvses -= 1
                print("ses:",self.tvses)
            elif cevap == ">":
                self.tvses +=1
                print("ses:",self.tvses)
            else:
                print("Ses güncellendi")
                print("Ses:",self.tvses)
                break
    def kanal_ekle(self, yenikanal):
        print(f"Kanal ekleniyor: {yenikanal}")
        self.kanallistesi.append(yenikanal)
        print(f"Mevcut kanallar: {self.kanallistesi}")
    def televizyon_bilgileri(self):
        print(f"""
        Televizyon Durumu: {self.tvdurum}
        Ses Seviyesi: {self.tvses}
        Mevcut Kanal: {self.kanal}
        Kanal Listesi: {self.kanallistesi}
        """)
kumanda=kumanda()
print("""
Televizyon Kumandası Uygulamasına Hoş Geldiniz!

1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Televizyon Bilgileri

Çıkmak için 'q' tuşuna basın.
    """)
while True:
        islem = input("Yapmak istediğiniz işlemi seçin: ")

        if islem == 'q':
            print("Programdan çıkılıyor...")
            time.sleep(1)
            break
        elif islem == '1':
            kumanda.tvac()
        elif islem == '2':
            kumanda.tvkapat()
        elif islem == '3':
            kumanda.sesayarları()
        elif islem == '4':
            kanal_isim = input("Eklemek istediğiniz kanalın ismi: ")
            kumanda.kanal_ekle(kanal_isim)
        elif islem == '5':
            kumanda.televizyon_bilgileri()
        else:
            print("Geçersiz işlem. Lütfen tekrar deneyin.")





```
